```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

from util import visualize_classifier
```

```python
X, y = make_blobs(n_samples=120,
                  centers=[[0,0], [0,2], [-2,1]],
                  random_state=123,
                  cluster_std=0.6) 

plt.scatter(X[:, 0], X[:, 1], c=y, s=50)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X, y)

visualize_classifier(knn, X, y)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

tree = DecisionTreeClassifier()
tree.fit(X, y)

visualize_classifier(tree, X, y)
```

```python
plt.figure(figsize=(10, 8))
plot_tree(tree, feature_names=["x1", "x2"], class_names=["t1", "t2", "t3"], filled=True)
plt.show()
```

```python
X, y = make_blobs(n_samples=500, 
                  centers=[[0,0], [0,2], [-2,1], [-2,2], [3,3], [1,-2]],
                  random_state=123,
                  cluster_std=0.6)
```

```python
fig, axs = plt.subplots(3, 4, figsize=(12, 12))

for depth, ax in enumerate(axs.flat):
    model = DecisionTreeClassifier(max_depth=depth+1)
    model.fit(X, y)
    visualize_classifier(model, X, y, ax=ax, title=f"depth {depth}")

plt.show()
```
